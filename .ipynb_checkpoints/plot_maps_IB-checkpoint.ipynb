```python
%matplotlib inline

import matplotlib.pyplot as plt
import geopandas as gpd
import palettable as pltt
from seaborn import palplot
```

```python
fname_shp = './data/Municipios_IGN/Municipios_IGN.shp'
gdb_IB = gpd.read_file(fname_shp)
```

```python
gdb_IB.head()
```

```python
gdb_IB.plot()
```

```python
gdb_IB.loc[0, 'geometry']
```

```python
# Setup figure and axis
f, ax = plt.subplots(1, figsize=(12, 6))
# Plot layer of polygons on the axis
gdb_IB.plot(ax=ax)
# Remove axis frames
ax.set_axis_off()
# Force axis to be on the same unit
plt.axis('equal')
# Add figure title
f.suptitle('Balearic Islands')
# Display
plt.show()
```
